```python
import folium
```

```python
m = folium.Map(location=[46.204391,6.143158], tiles="CartoDB Positron", zoom_start=1.5)
m
```

```python
folium.Marker(location=[46.204391,6.143158], popup="<strong>CERN</strong>", tooltip="Clock for more information").add_to(m)
m
```

```python
folium.Marker(location=[37.774929,-122.419418], popup="<strong>San Francisco</strong>", tooltip="Clock for more", icon=folium.Icon(icon='cloud')).add_to(m)
m
```

```python
folium.Marker(location=[37.774929,-122.419418], popup="<strong>San Francisco</strong>", tooltip="Clock for more", icon=folium.Icon(icon='cloud', color='red')).add_to(m)
folium.Marker(location=[41.878113,-87.629799], popup="<strong>Chicago</strong>", tooltip="Clock for more", icon=folium.Icon(icon='envelope', color='green')).add_to(m)
m
```
